```python
import numpy as np
import matplotlib.pyplot as plt

#1
a = np.random.rand(100)
b = np.random.rand(100)
plt.scatter(a, b)
plt.show()

#2
mean = np.array([0, 1])
cov = np.array([[1, 0.25], [0.5, 1]])
x, y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(x, y, 'go')
plt.axis('equal')
plt.show()
```
