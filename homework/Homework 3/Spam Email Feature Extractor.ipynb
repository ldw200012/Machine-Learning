```python
import numpy as np

# ASCII CODE TABLE
# Beginning: 0
# 48-57 : Integer
# 65-90: Capital Letter
# 97-122: Low Letter
# End: 127

ascii_range = list(range(128))
integer_range = list(range(48, 58))
capitalletter_range = list(range(65, 91))
lowletter_range = list(range(97, 123))
non_alphanumerical = list()

for i in ascii_range:
    if i not in integer_range and i not in capitalletter_range and i not in lowletter_range:
        non_alphanumerical.append(chr(i))
        


keyword_list = list(["make", "address", "all" , "3d", "our", "over", "remove", "internet",
                    "order", "mail", "receive", "will", "people", "report", "addresses", "free",
                    "business", "email", "you" , "credit", "your", "font", "000", "money", "hp",
                    "hpl", "george", "650", "lab", "labs", "telnet", "857", "data", "415", "85",
                    "technology", "1999", "parts", "pm", "direct", "cs", "meeting", "original",
                    "project", "re", "edu", "table", "conference"])



keycharacter_list = list([';', '(', '[', '!', '$', '#'])

# print(non_alphanumerical)
# print(keyword_list)
# print(keycharacter_list)
```

```python

# Extracting Features Function

def spam_or_not(email_name):
    email_1 = open(email_name, "r").read()
    email_1_modified = email_1

    # Get List of words from Email_1
    for index, char in enumerate(email_1_modified):
        if char in non_alphanumerical:
            email_1_modified = email_1_modified.replace(char, " ")
    email_1_words = email_1_modified.split()

    # email_1_words_1
    # ---------------------------- print("Email 1 : ", email_1_words, '\n', sep='\n') ---------------------------- 
    
    # Extract Required Variables

    # Percentage of 48 words
    # Model Keywords
    # make, address, all , 3d, our, over, remove, internet,
    # order, mail, receive, will, people, report, addresses,
    # free, business, email, you , credit, your, font, 000, 
    # money, hp, hpl, george, 650, lab, labs, telnet, 857, 
    # data, 415, 85, technology, 1999, parts, pm, direct, cs,
    # meeting, original, project, re, edu, table, conference

    output_word_list = np.zeros(48)

    # Count number of keywords in the email
    for word in email_1_words:
        for idx, keyword in enumerate(keyword_list):
            if word == keyword:
                output_word_list[idx] += 1

    # Get percentage of keywords in the email
    output_word_list = output_word_list/len(email_1_words)*100

    # ---------------------------- print(output_word_list) ----------------------------
    
    # Percentage of 6 characters
    # Model Keycharacters
    # ;, (, [, !, $, #

    output_character_list = np.zeros(6)

    # Count number of keycharacters in the email
    for character in email_1:
        for idx, keycharacter in enumerate(keycharacter_list):
            if character == keycharacter:
                output_character_list[idx] += 1

    # Get percentage of keycharacters in the email
    output_character_list = output_character_list/len(email_1)*100       

    # ---------------------------- print(output_character_list) ----------------------------
    
    # Capital length list of each word in the email

    capital_length_in_email_list = list()

    for word in email_1_words:
        capital_count_list = list()
        capital_count = 0
        for char in word:
            if ord(char) in capitalletter_range:
                capital_count += 1
            else:
                capital_count_list.append(capital_count)
                capital_count = 0

            capital_count_list.append(capital_count)

        capital_length = max(capital_count_list)    
        capital_length_in_email_list.append(capital_length)

    # Average Capital Length

    average_capital_length = sum(capital_length_in_email_list) / len(capital_length_in_email_list)

    # Longest Capital Length

    longest_capital_length = max(capital_length_in_email_list)

    # Total Capital Length

    total_capital_length = sum(capital_length_in_email_list)

    # ------------------------ print(average_capital_length, longest_capital_length, total_capital_length) ------------------------
    
    result = np.array([])
    result = np.append(result,output_word_list)
    result = np.append(result,output_character_list)
    result = np.append(result,average_capital_length)
    result = np.append(result,longest_capital_length)
    result = np.append(result,total_capital_length)
    result = result.flatten()

    return result
```

```python
print("Email 1 : ", spam_or_not("spam_or_no_spam_1.txt"), '\n', sep='\n')
print("Email 2 : ", spam_or_not("spam_or_no_spam_2.txt"), '\n', sep='\n')
print("Email 3 : ", spam_or_not("spam_or_no_spam_3.txt"), '\n', sep='\n')
```
